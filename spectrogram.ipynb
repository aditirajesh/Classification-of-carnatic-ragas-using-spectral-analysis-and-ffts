```python
import numpy as np
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
```

```python
sample_rate, audio_data = wavfile.read('Recordings/001.Kanakangi/3 kanakAngi_freq40_4.wav')
```

```python
# If stereo, convert to mono by averaging channels
if len(audio_data.shape) > 1:
    audio_data = np.mean(audio_data, axis=1)
```

```python
# Normalize the amplitude of the signal for better plotting (optional)
audio_data = audio_data / np.max(np.abs(audio_data))
```

```python
time_axis = np.linspace(0, len(audio_data) / sample_rate, num=len(audio_data))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_axis, audio_data)
plt.title('Audio Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')


plt.show()
```

```python
from scipy.fft import fft, fftfreq
yf = fft(audio_data)
xf = fftfreq(len(audio_data), 1 / sample_rate)
plt.plot(xf, np.abs(yf))
plt.show()
```
